```python

# Load the SMAP L3 dataset (surface soil moisture data)
smap_dataset = ee.ImageCollection('NASA_USDA/HSL/SMAP_soil_moisture')

# Filter the dataset by date (e.g., from April 1, 2015, to April 2, 2024)
start_date = '2015-04-01'
end_date = '2024-04-02'
smap_image = smap_dataset.filterDate(start_date, end_date).first()

# Inspect the available bands in the image
band_names = smap_image.bandNames().getInfo()
print('Available bands:', band_names)

# Select the specific soil moisture profile band
soil_moisture_data_ssm = smap_image.select('ssm')
soil_moisture_data_susm = smap_image.select('susm')
soil_moisture_data_smp = smap_image.select('smp')
soil_moisture_data_ssma = smap_image.select('ssma')
soil_moisture_data_susma = smap_image.select('susma')

# Load the country boundaries from the USDOS LSIB dataset
countries = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017')

# Filter to get the boundaries of India
india_boundary = countries.filter(ee.Filter.eq('country_na', 'India'))

# If you want to include Sri Lanka, filter for Sri Lanka as well
sri_lanka_boundary = countries.filter(ee.Filter.eq('country_na', 'Sri Lanka'))

# Merge India and Sri Lanka boundaries into one feature collection
india_and_sri_lanka = india_boundary.merge(sri_lanka_boundary)

# Clip the SMAP data to the India and Sri Lanka boundaries
soil_moisture_clipped_ssm = soil_moisture_data_ssm.clip(india_and_sri_lanka)
soil_moisture_clipped_susm = soil_moisture_data_susm.clip(india_and_sri_lanka)
soil_moisture_clipped_smp = soil_moisture_data_smp.clip(india_and_sri_lanka)
soil_moisture_clipped_ssma = soil_moisture_data_ssma.clip(india_and_sri_lanka)
soil_moisture_clipped_susma = soil_moisture_data_susma.clip(india_and_sri_lanka)

# Set visualization parameters for soil moisture
visualization_params = {
    'min': 0.0,
    'max': 0.6,
    'palette': ['fc6238', 'ffec59', '8dd7bf', '00b0ba', '0065a2']
}

# Use geemap to display the map centered over India and Sri Lanka
Map = geemap.Map(center=[20.5937, 78.9629], zoom=5)  # Center over India
Map.addLayer(soil_moisture_clipped_ssm, visualization_params, 'SSM')
Map.addLayer(soil_moisture_clipped_susm, visualization_params, 'susm')
Map.addLayer(soil_moisture_clipped_smp, visualization_params, 'SMP')
Map.addLayer(soil_moisture_clipped_ssma, visualization_params, 'ssma')
Map.addLayer(soil_moisture_clipped_susma, visualization_params, 'susma')

# Display the map
Map
```

```python
import ee
import geemap
from datetime import datetime, timedelta

# Initialize the Earth Engine API
ee.Initialize()

# Load the SMAP L3 dataset (surface soil moisture data)
smap_dataset = ee.ImageCollection('NASA_USDA/HSL/SMAP_soil_moisture')

# Define the start and end dates for looping
start_date = datetime.strptime('2019-08-01', '%Y-%m-%d')
end_date = datetime.strptime('2024-10-01', '%Y-%m-%d')  # Set an example end date for a short loop

# Load the country boundaries from the USDOS LSIB dataset
countries = ee.FeatureCollection('USDOS/LSIB_SIMPLE/2017')
india_boundary = countries.filter(ee.Filter.eq('country_na', 'India'))
sri_lanka_boundary = countries.filter(ee.Filter.eq('country_na', 'Sri Lanka'))
india_and_sri_lanka = india_boundary.merge(sri_lanka_boundary)

# Set visualization parameters for soil moisture
visualization_params = {
    'min': 0.0,
    'max': 0.6,
    'palette': ['fc6238', 'ffec59', '8dd7bf', '00b0ba', '0065a2']
}

# Loop through each date
current_date = start_date
while current_date <= end_date:
    # Convert the date to string in required format
    date_str = current_date.strftime('%Y-%m-%d')
    next_date = current_date + timedelta(days=1)
    next_date_str = next_date.strftime('%Y-%m-%d')

    # Filter the dataset for the current date
    smap_image = smap_dataset.filterDate(date_str, next_date_str).first()

    # Handle missing data by skipping
    try:
        # Check if the image is valid (not None)
        if smap_image.getInfo() is None:
            raise ValueError(f"No data found for date {date_str}\n")

        # Check if the image has any bands
        band_names = smap_image.bandNames().getInfo()
        if not band_names:
            raise ValueError(f"No bands found for date {date_str}\n")

        # Select the specific soil moisture profile band
        soil_moisture_data_smp = smap_image.select('smp')

        # Clip the SMAP data to the India and Sri Lanka boundaries
        soil_moisture_clipped = soil_moisture_data_smp.clip(india_and_sri_lanka)

        # Use geemap to display the map centered over India and Sri Lanka
        Map = geemap.Map(center=[20.5937, 78.9629], zoom=5)
        Map.addLayer(soil_moisture_clipped, visualization_params, f'Soil Moisture - {date_str}')
        print(f"Displaying soil moisture for {date_str}")
        Map  # This will display the map for each date iteration if running interactively (e.g., in Jupyter Notebook)

        # Define a region of interest covering India and Sri Lanka
        roi = india_and_sri_lanka.geometry()

        # Export the image to a specific folder in Google Drive
        export_task = ee.batch.Export.image.toDrive(
            image=soil_moisture_clipped,
            description=f'soil_moisture_india_and_sri_lanka_{date_str}',  # Unique description for each date
            folder='NASA_Soil_Moisture_India_Dataset',  # Specify the Google Drive folder name here
            scale=9000,  # Set resolution in meters
            region=roi,  # Set the region of interest to India and Sri Lanka boundary
            fileFormat='GeoTIFF',
            maxPixels=1e10  # Allow for large pixel exports
        )

        # Start the export task
        export_task.start()
        print(f"Export task started for {date_str}. Check your Google Drive 'NASA_Soil_Moisture_India_Dataset' folder for the file once completed.\n")

    except ValueError as e:
        print(e)  # Print the error message for missing data and skip to the next date

    # Move to the next date
    current_date = next_date
```
